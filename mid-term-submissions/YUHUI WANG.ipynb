```python
c1=[]
for i in range(len(v1)):
    if v1.values[i][0]=='Yes':
        c1.append('Developed region')
    elif v1.values[i][1]=='Yes':
        c1.append('Least developed country')
    elif v1.values[i][2]=='Yes':
        c1.append('Sub-Saharan Africa')
    else :
        c1.append('Less developed region')
c1
```
